## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
charity_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
charity_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = charity_df.drop(columns=["EIN", "NAME"]) # keeping names to determine effect on model performance
application_df.head()
```

```python
# Determine the number of unique values in each column.
application_df.nunique(axis=0)
```

```python
# Look at APPLICATION_TYPE value counts for binning
app_counts = application_df["APPLICATION_TYPE"].value_counts()
app_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = list(app_counts[app_counts<528].index)
# application_types_to_replace # to check for values in list

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
class_counts = application_df["CLASSIFICATION"].value_counts()
class_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_counts[class_counts>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
classifications_to_replace = list(class_counts[class_counts<1883].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
numeric_df = pd.get_dummies(application_df)
numeric_df.head()
```

```python
from pandas.core.arrays import numeric
# Split our preprocessed data into our features and target arrays
y = numeric_df['IS_SUCCESSFUL']
X = numeric_df.drop(columns='IS_SUCCESSFUL')

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
X_train_scaled.shape
```

## Compile, Train and Evaluate the Model

Model 1

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(input_dim=43, units=80, activation='relu')
)
# Second hidden layer
nn.add(
    tf.keras.layers.Dense(input_dim=80, units=30, activation='relu')
)
# Output layer
nn.add(
    tf.keras.layers.Dense(units=1, activation='sigmoid')
)
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train the model
nn.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

Model 2 (added another hidden layer)

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn02 = tf.keras.models.Sequential()

# First hidden layer
nn02.add(
    tf.keras.layers.Dense(input_dim=43, units=80, activation='relu')
)
# Second hidden layer
nn02.add(
    tf.keras.layers.Dense(input_dim=80, units=30, activation='relu')
)
# Third hidden layer
nn02.add(
    tf.keras.layers.Dense(input_dim=30, units=15, activation='relu')
)
# Output layer
nn02.add(
    tf.keras.layers.Dense(units=1, activation='sigmoid')
)
# Check the structure of the model
nn02.summary()
```

```python
# Compile the model
nn02.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
nn02.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss02, model_accuracy02 = nn02.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss02}, Accuracy: {model_accuracy02}")
```
